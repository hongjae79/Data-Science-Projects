```python
def biggest(x, y, z):
    
    Max = x
    status = 'positive'
    if y > Max:
        Max = y   
        status = 'neutral'
    if z > Max:
        Max = z
        status = 'negative'
        if y > z:
            Max = y
            status = 'neutral'
    return Max, status
```
